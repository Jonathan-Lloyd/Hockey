```python
salaries['yearid'] = salaries['YEAR'].astype(str) + '-' + (salaries['YEAR'] + 1).astype(str)
salaries.info()
```

```python
salaries = salaries.reindex(columns=['PLAYER','CAP HIT','SALARY PERCENTAGE','yearid'])
salaries = salaries.rename(columns = {'PLAYER': 'name',
                                  'CAP HIT':'cap_hit',
                                  'SALARY PERCENTAGE' : 'cap_pct',
                                  'yearid':'yearid'})
```

```python
master_df_combine = stats_with_grades.merge(salaries,how='inner', on=['name','yearid'])
```
