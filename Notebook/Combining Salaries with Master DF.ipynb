```python
salaries= pd.read_csv('../Data/Player_Salaries.csv')
```

```python
salaries.head()
```

```python
salaries.info()
```

```python
salaries['yearid'] = salaries['YEAR'].astype(str) + '-' + (salaries['YEAR'] + 1).astype(str)
salaries.info()
```

```python
salaries = salaries.reindex(columns=['PLAYER','CAP HIT','SALARY PERCENTAGE','yearid'])
salaries = salaries.rename(columns = {'PLAYER': 'name',
                                  'CAP HIT':'cap_hit',
                                  'SALARY PERCENTAGE' : 'cap_pct',
                                  'yearid':'yearid'})
```

```python
master_df_combine = stats_with_grades.merge(salaries,how='left', on=['name','yearid'])
```

```python
master_df_combine[master_df_combine['yearid']== '2011-2012']
```

```python
master_df_combine.to_csv('../Data/stats_with_salaries.csv')
```

```python
standings = pd.read_csv('../Data/Team_Standings.csv')
```

```python
standings.head()
```

```python
new = standings['Season'].str.split('-',n = 1, expand = True)
new.info()
```

```python
standings['year']= new[0].astype(int)
standings.drop(columns =['Season'], inplace = True)
standings['yearid'] = standings['year'].astype(str) + '-' + (standings['year'] + 1).astype(str)
standings.info()
```

```python
standings.to_csv('../Data/teams_standings_updated.csv')
```
