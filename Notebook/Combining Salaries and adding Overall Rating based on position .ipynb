```python
salaries= pd.read_csv('../Data/Player_Salaries.csv')
```

```python
salaries.head()
```

```python
salaries.info()
```

```python
salaries['yearid'] = salaries['YEAR'].astype(str) + '-' + (salaries['YEAR'] + 1).astype(str)
salaries.info()
```

```python
salaries = salaries.reindex(columns=['PLAYER','CAP HIT','SALARY PERCENTAGE','yearid'])
salaries = salaries.rename(columns = {'PLAYER': 'name',
                                  'CAP HIT':'cap_hit',
                                  'SALARY PERCENTAGE' : 'cap_pct',
                                  'yearid':'yearid'})
```

```python
master_df_combine = stats_with_grades.merge(salaries,how='left', on=['name','yearid'])
```

```python
master_df_combine[master_df_combine['yearid']== '2011-2012']
```

```python
def score_rating(row):
    if (row['fwd_scoring_rtg']+row['fwd_pass_rtg']+row['fwd_shift_usage_rtg']) >= 27:
        val = 'Elite'
    elif (row['fwd_scoring_rtg']+row['fwd_pass_rtg']+row['fwd_shift_usage_rtg']) >= 24:
        val = 'Great'
    elif (row['fwd_scoring_rtg']+row['fwd_pass_rtg']+row['fwd_shift_usage_rtg']) >= 21:
        val = 'Above-Average'
    elif (row['fwd_scoring_rtg']+row['fwd_pass_rtg']+row['fwd_shift_usage_rtg']) >= 18:
        val = 'Average'
    elif (row['fwd_scoring_rtg']+row['fwd_pass_rtg']+row['fwd_shift_usage_rtg']) <= 17:
        val = 'Below Average'
    return val

master_df_combine['Forward_Skill'] = master_df_combine[master_df_combine['position']=='Forward'].apply(score_rating, axis=1)
master_df_combine
```

```python
def defense_rating(row):
    if (row['def_scoring_rtg']+row['def_pass_rtg']+row['def_shift_usage_rtg'])  >= 27:
        val = 'Elite'
    elif (row['def_scoring_rtg']+row['def_pass_rtg']+row['def_shift_usage_rtg']) >=24 :
        val = 'Great'
    elif (row['def_scoring_rtg']+row['def_pass_rtg']+row['def_shift_usage_rtg']) >= 21:
        val = 'Above-Average'
    elif (row['def_scoring_rtg']+row['def_pass_rtg']+row['def_shift_usage_rtg']) >= 18:
        val = 'Average'
    elif (row['def_scoring_rtg']+row['def_pass_rtg']+row['def_shift_usage_rtg']) <= 17:
        val = 'Below Average'
    return val

master_df_combine['Defense_Skill'] = master_df_combine[master_df_combine['position']=='Defenseman'].apply(defense_rating, axis=1)
master_df_combine
```

```python
def goalie_rating(row):
    if (row['goalie_usage_rating']+row['goalie_save_rating']+row['goalie_gaa_rtg']) >= 27:
        val = 'Elite'
    elif (row['goalie_usage_rating']+row['goalie_save_rating']+row['goalie_gaa_rtg']) >= 24:
        val = 'Great'
    elif (row['goalie_usage_rating']+row['goalie_save_rating']+row['goalie_gaa_rtg']) >= 21:
        val = 'Above-Average'
    elif (row['goalie_usage_rating']+row['goalie_save_rating']+row['goalie_gaa_rtg']) >= 18:
        val = 'Average'
    elif (row['goalie_usage_rating']+row['goalie_save_rating']+row['goalie_gaa_rtg']) <= 17:
        val = 'Below Average'
    return val

master_df_combine['Goalie_Skill'] = master_df_combine[master_df_combine['position']=='Goalie'].apply(goalie_rating, axis=1)
master_df_combine.info()
```

```python
master_df_combine['Skill'] = master_df_combine.bfill(axis=1).iloc[:,41]
master_df_combine
```

```python
master_df_combine.to_csv('../Data/stats_with_salaries.csv')
```
